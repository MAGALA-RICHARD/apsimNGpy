```python
from apsimNGpy.core.apsim import ApsimModel
#model = CompareCultivar(model='Maize')
import pandas as pd
pd.set_option('display.max_columns', None)
model = ApsimModel('Maize')
model.add_factor(specification="[Sow using a variable rule].Script.CultivarName = B_110, A_90")

model.create_experiment(permutation=True)
model.add_factor(specification="[Sow using a variable rule].Script.CultivarName =  Dekalb_XL82, Melkassa, Pioneer_34K77, Laila, B_110, A_90")
model.run()
df = model.results
    
```

```python
df['Year'] = [date.year for date in pd.to_datetime(df['Clock.Today'])]
```

```python
model.boxplot("Yield", by  = 'CultivarName')
```

Add another factor to check with cultivar withstands high population density



```python
model.add_factor(specification="[Sow using a variable rule].Script.Population =  4, 6, 8, 10, 12, 14")
model.run()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 10))
model.line_plot("Yield", x='Population', by  = 'CultivarName', estimator = 'mean',)
```
